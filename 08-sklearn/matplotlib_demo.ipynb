```python
import numpy as np
from matplotlib import pyplot as plt

if __name__ == "__main__":
    ys = 200 + np.random.randn(100)
    x = [x for x in range(len(ys))]

    plt.plot(x, ys, '-')
    plt.fill_between(x, ys, 195, where=(ys > 200), facecolor='g', alpha=0.6)

    plt.title("Sample Visualization")
    plt.show()
```
